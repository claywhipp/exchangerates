```python
import pandas as pd

# Read the CSV into a dataframe
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')

# Inspect the first 5 rows
exchange_rates.head()
```

```python
# Inspect the last 5 rows
exchange_rates.tail()
```

```python
# Get info about the columns
exchange_rates.info()
```

```python
# Rename, convert, sort certain columns
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
# Select only Time and US_dollar columns 
# Create a copy to avoid accidently changing the original dataframe
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
# Check to see if there any odd values (there are 62 counts of '-' when we expecto only numbers)
euro_to_dollar['US_dollar'].value_counts()
```

```python
# Remove rows with a US_dollar value of '-'
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'].str.contains('-') == False]

# Convert column to type float
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype('float')
```

```python
# Create a 30-day rolling mean of the US_dollar column
# A rolling mean is a mean that includes a given day and the 29 previous days. This means for the first 29 days
# the rolling mean will be blank since there's not enough data to calculate it for a 30 day window
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
# You can see rolling-mean has 29 fewer values than US_dollar
euro_to_dollar.info()
```

```python
# Select data between 2006 and 2009
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <= 2009)]

# Create 30-day rolling mean
financial_crisis['rolling_mean'] = financial_crisis['US_dollar'].rolling(30).mean()

# Select data between 2007 and 2008 
financial_crisis_78 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007) & (euro_to_dollar['Time'].dt.year <= 2008)]

# Create 30-day rolling mean
financial_crisis_78['rolling_mean'] = financial_crisis_78['US_dollar'].rolling(30).mean()
```

```python
# Import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# Use fivethirtyeight style
import matplotlib.style as style
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(8, 3))

# Plot data
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'], linewidth=1, color='lightgreen')
ax.plot(financial_crisis_78['Time'], financial_crisis_78['rolling_mean'], color='red', linewidth=3)

# Highlight peak of crisis
ax.bar(733205, height=0.4, width=190, bottom=1.2, alpha=0.5, color='darkgrey')

# Set x and y labels
ax.set_xticklabels([])
ax.set_yticklabels([])

y = 1.193
for rate in ['1.2', '1.3', '1.4', '1.5', '1.6']:
    ax.text(732172, y, rate, alpha=0.5, fontsize=11)
    y += 0.1
    
x = 732272
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.14, year, alpha=0.5, fontsize=11)
    x += 365
    
# Add title
ax.text(732172.0, 1.69,"Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis", weight='bold')
ax.text(732172.0, 1.65, "Euro-USD exchange rates between 2006 and 2010")

# Add signature
ax.text(732172.0, 1.07, ' '*122 + 'Source: European Central Bank', color = '#f0f0f0', backgroundcolor = '#4d4d4d', size=10)

# Make grid more transparent
ax.grid(alpha=0.5)

plt.show()
```
